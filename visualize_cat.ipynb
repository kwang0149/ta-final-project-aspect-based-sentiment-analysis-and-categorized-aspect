```python
import json

# Load data from JSONL file
data = []
with open('output_formatted.jsonl', 'r') as file:
    for line in file:
        data.append(json.loads(line))

# Count occurrences of each category
category_counts = {}
for entry in data:
    labels = entry.get('labels', [])
    for label in labels:
        if 'category' in label and label['category']:  # Check if 'category' exists and is not null
            main_category = label['category'].split('#')[0]
            if main_category not in category_counts:
                category_counts[main_category] = 0
            category_counts[main_category] += 1

# Handle null values
null_categories = 0
for entry in data:
    if not entry.get('labels'):
        null_categories += 1

if null_categories > 0:
    category_counts['NULL'] = null_categories

# Plotting
import matplotlib.pyplot as plt

categories = list(category_counts.keys())
counts = list(category_counts.values())

plt.figure(figsize=(10, 6))
plt.barh(categories, counts, color='skyblue')
plt.xlabel('Count')
plt.ylabel('Category')
plt.title('Aspect Categories Visualization')
plt.gca().invert_yaxis()
plt.show()
```
