```python
import json

# Load data from JSONL file
data = []
with open('output_formatted.jsonl', 'r') as file:
    for line in file:
        data.append(json.loads(line))

# Count occurrences of each category
category_counts = {}
for entry in data:
    labels = entry.get('labels', [])
    for label in labels:
        if 'category' in label and label['category']:  # Check if 'category' exists and is not null
            main_category = label['category'].split('#')[0]
            if main_category not in category_counts:
                category_counts[main_category] = 0
            category_counts[main_category] += 1

# Handle null values
null_categories = 0
for entry in data:
    if not entry.get('labels'):
        null_categories += 1

if null_categories > 0:
    category_counts['NULL'] = null_categories

# Plotting
import matplotlib.pyplot as plt

categories = list(category_counts.keys())
counts = list(category_counts.values())

plt.figure(figsize=(10, 6))
plt.barh(categories, counts, color='skyblue')
plt.xlabel('Count')
plt.ylabel('Category')
plt.title('Aspect Categories Visualization')
plt.gca().invert_yaxis()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Load data from JSONL file
data = []
with open('output_formatted.jsonl', 'r') as file:
    for line in file:
        data.append(json.loads(line))

# Extracting and counting occurrences of the second part of the 'LOCATION' category with polarity
hotel_context_counts = {}
for entry in data:
    labels = entry.get('labels', [])
    for label in labels:
        if 'category' in label and label['category']:
            main_category, context = label['category'].split('#')
            if main_category == 'LOCATION':
                if context not in hotel_context_counts:
                    hotel_context_counts[context] = {'POS': 0, 'NEG': 0}
                polarity = label['polarity']
                if polarity in hotel_context_counts[context]:
                    hotel_context_counts[context][polarity] += 1

# Prepare data for plotting
contexts = list(hotel_context_counts.keys())
positive_counts = [hotel_context_counts[context]['POS'] for context in contexts]
negative_counts = [hotel_context_counts[context]['NEG'] for context in contexts]

# Plotting
plt.figure(figsize=(10, 6))
bar_width = 0.4

# Create bars for positive and negative counts
bar1 = plt.barh(contexts, positive_counts, color='green', edgecolor='grey', height=bar_width, label='POS')
bar2 = plt.barh(contexts, negative_counts, left=positive_counts, color='red', edgecolor='grey', height=bar_width, label='NEG')

# Adding labels
for bar in bar1:
    width = bar.get_width()
    plt.text(width - 5, bar.get_y() + bar.get_height() / 2, f'{int(width)}', ha='center', va='center', color='white', fontweight='bold')
for bar in bar2:
    width = bar.get_width()
    plt.text(bar.get_x() + width / 2, bar.get_y() + bar.get_height() / 2, f'{int(width)}', ha='center', va='center', color='white', fontweight='bold')

plt.xlabel('Count')
plt.ylabel('Context')
plt.title('LOCATION Category Visualization with Polarity')
plt.legend()
plt.gca().invert_yaxis()
plt.show()
```
