```python
import json

def read_jsonl(file_path):
    with open(file_path, 'r') as file:
        data = []
        obj_str = ''
        for line in file:
            line = line.strip()
            if line:  # Skip empty lines
                obj_str += line
                try:
                    obj = json.loads(obj_str)
                    data.append(obj)
                    obj_str = ''  # Reset obj_str after successfully parsing an object
                except json.JSONDecodeError:
                    pass  # Incomplete JSON object, continue reading
    return data

def write_jsonl(data, file_path):
    with open(file_path, 'w') as file:
        for obj in data:
            json.dump({"index": obj["index"], **obj}, file)
            file.write('\n')

input_file = 'formatted_data_hotel_3.jsonl'
output_file = 'output_file_2.jsonl'

# Read the JSONL file and add index to each object
data = read_jsonl(input_file)
print(f"Read {len(data)} objects from {input_file}")

for idx, obj in enumerate(data, start=1):
    obj['index'] = idx

# Write the modified objects back to the JSONL file
write_jsonl(data, output_file)
print(f"Written {len(data)} objects to {output_file}")
```

```python
import json

def read_jsonl(file_path):
  """Reads a JSONL file and returns a list of dictionaries."""
  with open(file_path, 'r') as file:
    data = []
    index = 1  # Initialize index
    for line in file:
      line = line.strip()
      if line:  # Skip empty lines
        obj = json.loads(line)
        obj["index"] = index
        data.append(obj)
        index += 1  # Increment index for the next object
  return data

def format_label(label):
  """Returns a string representation of a label with proper indentation."""
  return json.dumps(label, indent=4)  # Use indentation for readability

def write_jsonl(data, file_path):
  """Writes a list of dictionaries to a new JSONL file with desired formatting."""
  with open(file_path, 'w', encoding='utf-8') as file:
    for obj in data:
      # Write "index" and "text" on separate lines with quotes
      file.write('{"index": "' + str(obj["index"]) + '",\n')
      file.write(f'"text": "{obj["text"]}",\n')

      # Write "labels" on a new line with opening bracket
      file.write('"labels": [\n')

      # Write each label with proper indentation and a comma (except the last)
      for i, label in enumerate(obj["labels"]):
        label_str = format_label(label)
        if i < len(obj["labels"]) - 1:  # Add comma for all but the last label
          label_str += ','
        file.write(f'{label_str}\n')

      # Write closing bracket for "labels" on a new line
      file.write(']\n')

      # Write closing curly brace for the object on a new line
      file.write("}\n")


input_file = 'output_file_2.jsonl'
output_file = 'output_file_3_indexed.jsonl'

# Read the JSONL file and extract data
data = read_jsonl(input_file)

# Write the modified objects to a new JSONL file
write_jsonl(data, output_file)

print(f"Data processed and written to {output_file} with desired formatting.")
```

```python
def extract_quadruples(data):
    extracted_data = []
    
    for entry in data:
        text = entry['data']['text']
        annotations = entry['annotations']
        
        labels = []
        
        if not annotations:
            continue
        
        results = annotations[0]['result']
        
        aspect_map = {}
        opinion_map = {}
        relations = []
        
        for item in results:
            item_value = item.get('value', {})
            item_labels = item_value.get('labels', [])
            item_type = item.get('type')
            
            if 'ASPECT' in item_labels:
                aspect_map[item['id']] = item_value.get('text', '')
            elif 'OPINION' in item_labels:
                opinion_map[item['id']] = (item_value.get('text', ''), item_value.get('meta', {}).get('text', ['neutral'])[0])
            elif item_type == 'relation':
                relations.append(item)
        
        for rel in relations:
            aspect_id = rel.get('from_id')
            opinion_id = rel.get('to_id')
            aspect = aspect_map.get(aspect_id)
            opinion, default_polarity = opinion_map.get(opinion_id, ('', 'neutral'))
            # Default values if 'labels' list is missing or doesn't have enough elements
            category = 'UNKNOWN'
            polarity = 'neutral'
            
            # Check if 'labels' list exists and has at least two elements
            if 'labels' in rel and len(rel['labels']) >= 2:
                polarity = rel['labels'][0].lower()  # First element for polarity
                category = rel['labels'][1]  # Second element for category
            elif 'labels' in rel and len(rel['labels']) == 1:
                # If there is only one element, use it for polarity
                polarity = rel['labels'][0].lower()
                
            if aspect and opinion:
                labels.append({
                    "aspect": aspect,
                    "opinion": opinion,
                    "polarity": polarity.lower(),
                    "category": category
                })
        
        if labels:
            extracted_data.append({
                "text": text,
                "labels": labels
            })
    
    return extracted_data
```
